```python
'''
 * User: Hojun Lim
 * Date: 2020-07-05
'''
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import matplotlib.image as mpimg
from PIL import Image 
import numpy as np
import os  

datasets = {'boston':[], 'concrete':[],'energy':[],'kin8nm':[],'power_plant':[],'wine':[],'yacht':[]
    ,'naval':[],'protein':[]#'year':[],
    }

OOD_output = {}
OOD_output_test = {}

for key in datasets.keys():
    OOD_output[key] = os.path.join('output_ood', key, 'figures')
    OOD_output_test[key] = os.path.join(OOD_output[key], 'test')
    
```

```python
def plot_GT_sub_Mean_figure_by_epochs(image_folder, key,  epoch):
    #'test-boston, GT-mean_and_std_of_PU_and_MC_epoch=150.png', 'test-boston, GT-mean_and_std_of_PU_and_MC_epoch=40.png', 'test-boston, GT-mean_and_std_of_PU_and_MC_with_ylim_epoch=150.png', 'test-boston, GT-mean_and_std_of_PU_and_MC_with_ylim_epoch=40.png', 'test-boston_NLL_RMSE_histogram_epoch=150.png', 'test-boston_NLL_RMSE_histogram_epoch=40.png'
    f, axes = plt.subplots(1,2, figsize=(50,50), tight_layout=True)
    img1 = mpimg.imread(os.path.join(image_folder, 'test-'+str(key)+', GT-mean_and_std_of_PU_and_MC_with_ylim_epoch='+str(epoch)+'.png'))
    img2 = mpimg.imread(os.path.join(image_folder, 'test-'+str(key)+', GT-mean_and_std_of_PU_and_MC_epoch='+str(epoch)+'.png'))
    
    axes[0].imshow(img1)
    axes[0].axis('off')
    
    axes[1].imshow(img2)
    axes[1].axis('off')
    plt.show()
    
    
def plot_NLL_figure(image_folder, key, epoch1, epoch2):
    f, axes = plt.subplots(1,2, figsize=(50,50), tight_layout=True)
    img1 = mpimg.imread(os.path.join(image_folder, 'test-'+str(key)+'_NLL_RMSE_histogram_epoch='+str(epoch1)+'.png'))
    img2 = mpimg.imread(os.path.join(image_folder, 'test-'+str(key)+'_NLL_RMSE_histogram_epoch='+str(epoch2)+'.png'))
    
    axes[0].imshow(img1)
    axes[0].axis('off')
    
    axes[1].imshow(img2)
    axes[1].axis('off')
    plt.show()
    
```

```python
for key in datasets.keys():
    plot_GT_sub_Mean_figure_by_epochs(OOD_output_test[key], key, 40)
    plot_GT_sub_Mean_figure_by_epochs(OOD_output_test[key], key, 150)
    plot_NLL_figure(OOD_output_test[key], key, 40,150)
    
    
```
